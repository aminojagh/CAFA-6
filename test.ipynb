```python
from cafaeval.evaluation import cafa_eval

result = cafa_eval(
    "data/example1/IDPO_disorder_function.obo",
    "data/example1/predictions",
    "data/example1/ground_truth.tsv"
)
```

The output of cafa_eval is a tuple containing:
- A pandas DataFrame with the evaluation results, one row per prediction file, namespace and threshold.
- A dictionary with the best scores (max F-measure, max Weighted F-measure, min Semantic similarity). For each score the dictionary contain a pandas DataFrame with one row per prediction file, namespace and threshold.
