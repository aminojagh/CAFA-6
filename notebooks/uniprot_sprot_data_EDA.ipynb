```python
import gzip
from Bio import SwissProt
from Bio import SeqIO
```

```python
counter = 0
with gzip.open("uniprot_sprot.dat.gz", "rt") as handle:
    for record in SwissProt.parse(handle):
        print(f"ID: {record.entry_name}, Accession: {record.accessions[0]}")
        counter+=1
        if counter >=10: break
```

```python
counter = 0
with gzip.open("uniprot_sprot.dat.gz", "rt") as handle:
    for record in SeqIO.parse(handle, "swiss"):
        print(record.id, record.seq)
        counter+=1
        if counter >=10: break
```

```python
counter = 0
with gzip.open("uniprot_sprot.fasta.gz", "rt") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        print(f"ID: {record.id}")
        print(f"Name: {record.name}")
        print(f"Sequence Length: {len(record.seq)}")
        counter+=1
        if counter >=10: break
```

```python
counter = 0
# Open the gzipped Swiss-Prot file
with gzip.open("uniprot_sprot.dat.gz", "rt") as handle:
    for record in SwissProt.parse(handle):
        # Extract GO terms
        go_terms = [ref for ref in record.cross_references if ref[0] == 'GO']
        print(f"ID: {record.accessions[0]}")
        print(f"GO Terms: {go_terms}")
        counter+=1
        if counter >=10: break
```

```python
counter = 0
with gzip.open("uniprot_sprot.dat.gz", "rt") as file:
    for line in file:
        print(line.strip())
        counter+=1
        if counter >=1000: break
```
